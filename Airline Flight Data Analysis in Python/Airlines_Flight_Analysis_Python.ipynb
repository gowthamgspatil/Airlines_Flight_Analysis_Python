###  Project Overview

This Kaggle dataset includes detailed flight data—such as **airlines, origin and destination airports, scheduled and actual departure/arrival times, delays, cancellations, fares**, and more—covering a large volume of real-world flights.

Before analysis, common preprocessing steps include:

- Cleaning missing values  
- Standardizing time formats  
- Filtering relevant columns like **airline name, origin/destination, price, class**, and **days left to departure**


###  Objective and Exploratory Analysis

**Typical goals and analyses include:**

- **Frequency Analysis**: Counting flights per airline, and visualizing distributions of departure/arrival times and routes.
- **Price Variation**: Investigating how fares differ by airline, departure/arrival time, class (economy vs business), and booking lead time.
- **Route Pricing Patterns**: Studying price variability across specific source–destination pairs.
- **Last-minute Pricing**: Evaluating how ticket prices change when booked only 1–2 days before departure.

> These mirror many standard EDA questions often addressed by users working with similar flight datasets.


###  Key Tools & Techniques

**Python Libraries Used:**

- `pandas` – Data preparation and cleaning  
- `seaborn` / `matplotlib` – Visualizations (bar plots, box plots, line charts)  
- `scipy.stats` – Statistical hypothesis testing  
- `scikit-learn` – Regression and classification modeling  

---

**Visualizations:**

- **Bar Charts**: For frequency analysis by airline, time of day, and routes  
- **Line Plots / Box Plots**: To show pricing patterns over time or across categories (e.g. departure vs arrival times, class types)  

---

**Statistical Testing:**

- Identify if price differences between classes (e.g. Economy vs Business) or days-before-departure are statistically significant

---

**Predictive Modeling:**

- Apply regression/classification models to predict ticket prices or causes of delays based on flight attributes

### Q1. Airlines & Frequencies:
#### Counted each airline’s occurrences in the dataset.

```python
data.head()
```

```python
data['airline'].nunique()
```

```python
data['airline'].unique()
```

```python
data['airline'].value_counts()
```

```python
airline_counts = data['airline'].value_counts(ascending=True)
colors = plt.cm.Set3(range(len(airline_counts)))  # You can also try: Set1, tab10, Accent, etc.

airline_counts.plot.barh(color=colors)
plt.title("Airlines with Frequencies")
plt.xlabel("Number of Flights")
plt.ylabel("Airlines")
plt.show()
```

### Q2. Departure & Arrival Time (Bar Graphs):
#### Plotted bar charts showing frequency of different departure and arrival times.

```python
data.head()
```

```python
data['departure_time'].value_counts()
```

```python
data['arrival_time'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Get counts
departure_counts = data['departure_time'].value_counts()
arrival_counts = data['arrival_time'].value_counts()

# Create color arrays
departure_colors = plt.cm.Set2(range(len(departure_counts)))
arrival_colors = plt.cm.Paired(range(len(arrival_counts)))

plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.bar(departure_counts.index, departure_counts.values, color=departure_colors)
plt.title("Departure Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")

plt.subplot(1, 2, 2)
plt.bar(arrival_counts.index, arrival_counts.values, color=arrival_colors)
plt.title("Arrival Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")

plt.tight_layout()
plt.show()
```

### Q3. Source & Destination Cities (Bar Graphs):
#### Visualized the number of flights per source and destination city.

```python
data.head()
```

```python
data['source_city'].value_counts()
```

```python
data['destination_city'].value_counts()
```

```python
source_counts = data['source_city'].value_counts()
dest_counts = data['destination_city'].value_counts()

source_colors = plt.cm.Set3(range(len(source_counts)))
dest_colors = plt.cm.Pastel1(range(len(dest_counts)))

plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.barh(source_counts.index, source_counts.values, color=source_colors)
plt.title("Source Cities with No. of Flights")
plt.ylabel("Cities")
plt.xlabel("No. of Flights")

plt.subplot(1, 2, 2)
plt.barh(dest_counts.index, dest_counts.values, color=dest_colors)
plt.title("Destination Cities with No. of Flights")
plt.ylabel("Cities")
plt.xlabel("No. of Flights")

plt.tight_layout()
plt.show()
```

### Q4. Price vs Airlines:
#### Analyzed how ticket prices vary across airlines.

```python
data.head()
```

```python
data.groupby('airline')['price'].mean()
```

```python
sns.catplot(
    x='airline',
    y='price',
    kind='bar',
    hue='class',
    data=data,
    palette='Set2'
)

plt.title("Average Price by Airline and Class")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
